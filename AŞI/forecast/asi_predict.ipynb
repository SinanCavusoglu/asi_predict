```python
import pandas as pd
import copy
```

```python
asi_dict = {"Hepatit B": (0,1,6),
            "BCG": (2,),         # (Verem)
            "DaBT-IPA-Hib": (2,4,6,18), # 4'lü Karma Aşı
            "KPA": (2,4,12),
            "KKK": (9,12,48),
            "DaBT-IPA": (48,),
            "OPA": (6,18),
            "Hepatit A": (18,24),
            "Suçiçeği": (12,)} 
```

```python
month_birth_df = pd.read_csv("dataset/full_montly_df.csv", index_col="Index")
month_birth_df.index = copy.deepcopy(pd.to_datetime(month_birth_df.index, format="%Y-%m-%d")) 
cities = month_birth_df.columns
```

```python
month_birth_df.iloc[48:]
```

```python
total_asi_index = month_birth_df.loc[pd.to_datetime("2013-01-01", format="%Y-%m-%d"):].index
total_asi_df = pd.DataFrame(index=total_asi_index)
```

```python
asi_dict
```

```python
#pd.DateOffset(months=3)
for city in cities:
    for date in total_asi_index:
        for asi in asi_dict.keys():
            patterns = asi_dict[asi]
            total = 0
            for pattern in patterns:
                total += month_birth_df.loc[date - pd.DateOffset(months = pattern), city]
            total_asi_df.loc[date, f"{city}_{asi}"] = total
```

```python
total_asi_df = total_asi_df.astype(int)
```

```python
total_asi_df.to_csv("dataset/predict_vaccine_montly.csv")
```

```python
total_asi_df_yearly = total_asi_df.resample("Y").sum()
```

```python
total_asi_df_yearly.to_csv("dataset/predict_vaccine_yearly.csv")
```
